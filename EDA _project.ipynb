<b style="font-size: 36px;">Data Visualisation</b>

```python
df_count=pd.DataFrame({"No_of_Employees":df.Team.value_counts()}).reset_index()
df_count.columns = ["Team","No_of_Employees"]
df_count=df_count[["Team","No_of_Employees"]]
df_count
```

```python
plt.figure(figsize=(15,10))
colors = plt.cm.viridis(np.linspace(0, 1, len(df_count)))
plt.bar(df_count["Team"],df_count["No_of_Employees"],color=colors)
plt.xticks(rotation="vertical")
plt.title("Number of Employees in each Team")
plt.xlabel("Teams")
plt.ylabel("Number of employees")
plt.show()
```

```python
df_count["Percentage"]=(df_count["No_of_Employees"] / df_count["No_of_Employees"].sum()) * 100
df_count
```

```python
plt.figure(figsize=(14,14))
plt.pie(df_count["Percentage"],labels=df_count["Team"], autopct='%1.1f%%')
plt.title("Employee percentage in each team",fontdict={'fontweight': 'bold'})
plt.show()
```

**Data Story**

Here we created a data frame which shows number of empoyees and we made a barplot with that.We added another column named percentage and made a pie chart which shows percentage of employeesin each team
